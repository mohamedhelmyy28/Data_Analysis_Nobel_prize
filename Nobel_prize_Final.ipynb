```python
most_win=nobel_df.groupby('full_name').count().reset_index()
most_win=most_win.sort_values(by='year', ascending=False).head(6).sort_values(by='year', ascending=True)
most_win=most_win[most_win['birth_date']!=0]
most_win
```

```python
cat_df=nobel_df.groupby('category').count().reset_index()
all_fields=nobel_df.groupby('year').count().reset_index()

chemist=nobel_df[nobel_df['category']=='Chemistry'].groupby('year').count().cumsum().reset_index()
Physics=nobel_df[nobel_df['category']=='Physics'].groupby('year').count().cumsum().reset_index()
Medicine=nobel_df[nobel_df['category']=='Medicine'].groupby('year').count().cumsum().reset_index()
Literature=nobel_df[nobel_df['category']=='Literature'].groupby('year').count().cumsum().reset_index()
Peace=nobel_df[nobel_df['category']=='Peace'].groupby('year').count().cumsum().reset_index()
Economics=nobel_df[nobel_df['category']=='Economics'].groupby('year').count().cumsum().reset_index()
```

```python
nobel_df['birth_date']=pd.to_datetime(nobel_df['birth_date'])

nobel_df['birth_year'] = nobel_df['birth_date'].dt.year
nobel_df['win_age']=nobel_df['year']-nobel_df['birth_year']
```

# Bar_Chart for countries

```python
bar_df=nobel_df.groupby('birth_country').count().reset_index()
bar_df=bar_df.sort_values(by='laureate_id', ascending=False).head(20).sort_values(by='laureate_id', ascending=True)

data_canada = px.data.gapminder().query("country == 'Canada'")
fig11 = px.bar(bar_df, x='laureate_id', y='birth_country',text=[str(i) for i in bar_df['laureate_id']]
            ,orientation='h', labels={
                     "laureate_id": "Count",
                     "birth_country": "Country"
                 })
fig11.show()
```

# Bar_Chart for organization

```python
bar_df=nobel_df.groupby('organization_name').count().reset_index()
bar_df=bar_df.sort_values(by='laureate_id', ascending=False).head(20).sort_values(by='laureate_id', ascending=True)

fig12 = px.bar(bar_df, x='laureate_id', y='organization_name',text=[str(i) for i in bar_df['laureate_id']]
            ,orientation='h', labels={
                     "laureate_id": "Count",
                     "organization_name": "Organization"
                 })
fig12.show()
bar_df.head()
```

```python
app = dash.Dash(__name__,external_stylesheets=['https://codepen.io/chriddyp/pen/bWLwgP.css'])




Total_Winners = go.Figure(go.Indicator(
    mode = "number",
    value = total,
    title = {"text": "Total Winners"},
    domain = {'row': 0, 'column': 0}))

indv_winners = go.Figure(go.Indicator(
    mode = "number",
    value = indv,
    number = {'prefix': ""},
    title = {"text": "Individual winners"},
    domain = {'row': 0, 'column': 1}))


org_Winners = go.Figure(go.Indicator(
    mode = "number",
    value = org,
    number = {'prefix': ""},
    title = {"text": "Organisations Winners"},
    domain = {'row': 0, 'column': 2}))

Winners_Alive = go.Figure(go.Indicator(
    mode = "number",
    value = 84+156,
    number = {'prefix': ""},
    title = {"text": "Alive"},
    domain = {'row': 0, 'column': 3}))




app.layout= html.Div([
    
    # Add a title
    html.H1(f'Nobel Prize ({begin}-{end})',
    style={'padding':'5px',"border":"2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'})
     ,html.Hr(style={'size':'10px'}),
    
     html.Div([ 
        html.Div([
            html.H2('Total Winners',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'}),
            html.H3(total,style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'}),
            
        ],className='three columns'), 
        
        
        html.Div([
            html.H2('Individual',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'}),
            html.H3(indv,style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'}),
            
        ],className='three columns'),
         
         html.Div([
            html.H2('Organization',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'}),
            html.H3(org,style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'}),
            
        ],className='three columns'),
         
         html.Div([
            html.H2('Alive',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'}),
            
            html.H3(240,style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':40 ,'background-color':'rgb(42, 1, 52)'}),
            
        ],className='three columns'),
       
        
                
        
    ],className='row'),
    
      html.Hr(style={'size':'10px'}),
        
      
    
    # 2 Charts
     html.Div([ 
        html.Div([
            html.H1('Total prizes per year',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':30 ,'background-color':'rgb(42, 1, 52)'}),
            dcc.Graph(figure= fig31),
        ],className='six columns'), 
        
        
        html.Div([
            html.H1('Male Vs Female',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':30 ,'background-color':'rgb(42, 1, 52)'}),
            dcc.Graph(figure= fig22),
        ],className='six columns')        
        
        
                
        
    ],className='row'),
    # 2 graphs 
       html.Div([ 
        html.Div([
            html.H1('Cumulative prizes per category',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':30 ,'background-color':'rgb(42, 1, 52)'}),
            dcc.Graph(figure= fig1),
        ],className='six columns'),
        
                     
        
        
        html.Div([
            html.H1('Total prizes per country',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':30 ,'background-color':'rgb(42, 1, 52)'}),
            dcc.Graph(figure= fig11),
        ],className='six columns')        
        
    ],className='row'),
       
    
    
    # 2 bars
     html.Div([ 
        
        
        
        
        html.Div([
            html.H1('Age distribution ',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':30 ,'background-color':'rgb(42, 1, 52)'}),
            dcc.Dropdown(
                id='dropdown',    
                options=[{'label': i, 'value': i} for i in nobel_df['category'].unique()],
                value=None,
                 style= {              
                       
 
                      'width': '100%',                  
                    'font-size': '20px'          
                    }
            ),
            dcc.Graph(figure= fig21,id='fig21'),
            
        ],className='six columns') ,
        
    
        html.Div([
            html.H1('Total prizes per organization',style={'padding':'0.5px',"border":"0.2px black solid",'color':'white','text-align':'center', 'font-size':30 ,'background-color':'rgb(42, 1, 52)'}),
            dcc.Graph(figure= fig12),
        ],className='six columns')      
        
    ],className='row'),


     ])


@app.callback(
    Output(component_id='fig21',component_property= 'figure'),
    Input(component_id='dropdown', component_property='value')
    )

def update(dropdown):
    if  dropdown==None or dropdown==[]:
        in_df = nobel_df.groupby('win_age').count().reset_index()
        

    else:
        filtered_df = nobel_df.groupby(['category','win_age']).count().reset_index()
        in_df=filtered_df[filtered_df['category']==dropdown]
        
    
    fig21 = px.bar(in_df, x='win_age', y='laureate_id',text=[str(i) for i in in_df['laureate_id']], labels={'win_age': 'Age', 'laureate_id':'Count'})
    
    return fig21

    

app.run_server()
```
